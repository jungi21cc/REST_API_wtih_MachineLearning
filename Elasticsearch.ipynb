# Elasticsearch

```python
from elasticsearch import Elasticsearch
esclient = Elasticsearch(hosts='localhost:9200')
```

```python
esclient.index(index='basketball', 
               doc_type='record', 
               body={"team" : "Chicago Bulls",
                     "name" : "Michael Jordan", 
                     "points" : 30,
                     "rebounds" : 3,
                     "assists" : 4, 
                     "submit_date" : "1996-10-11"},
               id=999
              )
```

```python
esclient.search(index='basketball', 
                doc_type='record', 
                body={ "query" : {"term" : { "points" : "30" }}}
               )
```

# search

```python
from pprint import pprint
###query DSL term
response = esclient.search(index='basketball', 
                           doc_type='record', 
                           body={ "query" : {"term" : { "rebounds" : 5}}}
                          )
pprint(response['hits']['hits'])
```

```python
###query DSL match
response = esclient.search(index='basketball', 
                           doc_type='record', 
                           body={ "query" : {"match" : { "team" : 'golden State Warrir'}}}
                          )
pprint(response['hits']['hits'])
```

```python
###query DSL range
response = esclient.search(index='basketball', 
                           doc_type='record', 
                           body={ "query" : {"range" : { "points" : {"gte" : 20,
                                                                       "lte" : 30,
                                                                       "boost" : 25.0}}}}
                          )
pprint(response['hits']['hits'])
```
